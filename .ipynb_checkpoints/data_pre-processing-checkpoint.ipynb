```python
import pandas as pd
```

```python
df = pd.read_csv('data/Space_Corrected.csv')
df
```

```python
# rename column Rocket to Cost (in million USD)
rocket_to_cost = {df.columns[7]: 'Cost'}
df.rename(columns = rocket_to_cost, inplace = True)

# remove the first two columns
t_data = df.drop(columns = ['Unnamed: 0', 'Unnamed: 0.1'])
t_data
```

```python
# Extract Country from Location and Year from Datum
t_data["Country"] = t_data["Location"].apply(lambda l: l.split(', ')[-1])

t_data["Year"] = pd.to_datetime(t_data["Datum"]).apply(lambda d: d.year)
```

```python
# Extract Rocket and Satellite info from Detail
t_data["Rocket"] = t_data["Detail"].apply(lambda rocket: rocket.split(' | ')[0])

t_data["Satellite"] = t_data["Detail"].apply(lambda rocket: rocket.split(' | ')[1])

t_data = t_data.drop(columns=['Detail'])
```

```python
clean_data = t_data
clean_data.to_csv(r'data/Space_Corrected_clean.csv', index = False)
```
