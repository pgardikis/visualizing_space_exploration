```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
exploration_data = pd.read_csv('data/Space_Corrected_clean.csv')
exploration_data
```

```python
# Number of launches per Company
plt.figure(figsize=(10,15))
axes = sns.countplot(y = 'Company Name', data = exploration_data, order = exploration_data['Company Name'].value_counts().index)
axes.set_xscale('log')
axes.set_title("Number of launches per Company", fontsize=18)
axes.set_ylabel("Company Name", fontsize=14)
axes.set_xlabel("Number of launches (log scale)", fontsize=14)

plt.show()
```

```python
# Number of launches per Country
plt.figure(figsize=(10,15))
axes = sns.countplot(y = 'Country', data = exploration_data, order = exploration_data['Country'].value_counts().index)
axes.set_xscale('log')
axes.set_title("Number of launches per Country", fontsize = 18)
axes.set_ylabel("Country", fontsize = 14)
axes.set_xlabel("Number of launches (log scale)", fontsize = 14)

plt.show()
```

```python
# Number of launches per Status Mission
plt.figure(figsize=(8,15))
axes = sns.countplot(x = 'Status Mission', data = exploration_data, order = exploration_data['Status Mission'].value_counts().index)
axes.set_yscale('log')
axes.set_title("Number of launches per Mission Status", fontsize=18)
axes.set_xlabel("Mission Status", fontsize = 14)
axes.set_ylabel("Number of launches (log scale)", fontsize = 14)

plt.show()
```

```python
# Number of launches per Year
plt.figure(figsize=(10,15))
axes = sns.countplot(y = 'Year', data = exploration_data)

axes.set_title("Launches per Year", fontsize = 18)
axes.set_ylabel("Year", fontsize = 14)
axes.set_xlabel("Number of Launches", fontsize = 14)

plt.show()
```

```python
# Number of launches per Rocket
plt.figure(figsize=(20,80))
axes = sns.countplot(y = 'Rocket', data = exploration_data, order = exploration_data['Rocket'].value_counts().index)
axes.set_xscale('log')
axes.set_title("Number of launches per Rocket", fontsize=18)
axes.set_ylabel("Rocket Name", fontsize=14)
axes.set_xlabel("Number of launches (log scale)", fontsize=14)

plt.show()
```

```python
# Number of launches per Rocket
plt.figure(figsize=(10,10))
axes = sns.countplot(x = 'Status Rocket', data = exploration_data, order = exploration_data['Status Rocket'].value_counts().index)
axes.set_title("Number of launches per Rocket Status", fontsize=18)
axes.set_xlabel("Rocket Status", fontsize=14)
axes.set_ylabel("Rockets", fontsize=14)

plt.show()
```
